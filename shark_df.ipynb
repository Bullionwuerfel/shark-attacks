```python
# Changing Column "Type" entries: Provoked (2 --> 1) & Questionable (2 --> 1) & Unverified =
shark_df["Type"]=shark_df["Type"].apply(lambda t: "Provoked" if t==" Provoked" else "Questionable" if t=="?" else  t)
```
